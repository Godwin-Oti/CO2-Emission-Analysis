# Fixing missing values

```python

# Calculate the mean of each column
#Total distance travelled in charge depleting operation with engine off (lifetime) (km)
column1_mean = car['Total distance travelled in charge depleting operation with engine off (lifetime) (km)'].mean()

# Fill missing values with the mean
car['Total distance travelled in charge depleting operation with engine off (lifetime) (km)'] = car['Total distance travelled in charge depleting operation with engine off (lifetime) (km)'].fillna(column1_mean)


#Total distance travelled in charge depleting operation with engine off (lifetime) (km)
column2_mean = car['Total distance travelled in charge depleting operation with engine running (lifetime) (km)'].mean()

# Fill missing values with the mean
car['Total distance travelled in charge depleting operation with engine running (lifetime) (km)'] = car['Total distance travelled in charge depleting operation with engine running (lifetime) (km)'].fillna(column2_mean)


#Total distance travelled in driver-selectable charge increasing operation (lifetime)  (km)
column3_mean = car['Total distance travelled in driver-selectable charge increasing operation (lifetime)  (km)'].mean()

# Fill missing values with the mean
car['Total distance travelled in driver-selectable charge increasing operation (lifetime)  (km)'] = car['Total distance travelled in driver-selectable charge increasing operation (lifetime)  (km)'].fillna(column3_mean)


#Total fuel consumed in charge depleting operation (lifetime) (l)
column4_mean = car['Total fuel consumed in charge depleting operation (lifetime) (l)'].mean()

# Fill missing values with the mean
car['Total fuel consumed in charge depleting operation (lifetime) (l)'] = car['Total fuel consumed in charge depleting operation (lifetime) (l)'].fillna(column4_mean)


#Total fuel consumed in driver-selectable charge increasing operation (lifetime) (l)
column5_mean = car['Total fuel consumed in driver-selectable charge increasing operation (lifetime) (l)'].mean()

# Fill missing values with the mean
car['Total fuel consumed in driver-selectable charge increasing operation (lifetime) (l)'] = car['Total fuel consumed in driver-selectable charge increasing operation (lifetime) (l)'].fillna(column5_mean)

#Total grid energy into the battery (lifetime) (kWh)
column6_mean = car['Total grid energy into the battery (lifetime) (kWh)'].mean()

# Fill missing values with the mean
car['Total grid energy into the battery (lifetime) (kWh)'] = car['Total grid energy into the battery (lifetime) (kWh)'].fillna(column6_mean)



```

Comparison of Fuel Consumption and CO2 Emissions

```python


# Group the data by 'Fuel Type' and 'Manufacturer' and calculate the average fuel consumption and CO2 emissions
average_data = car_agg.groupby(['Fuel Type', 'Manufacturer']).agg({'OBFCM Fuel consumption (l/100 km)': 'mean', 'OBFCM CO2 emissions (g/km)': 'mean'})

# Reset index to make the grouped columns accessible for analysis
average_data.reset_index(inplace=True)

# Print the average data
average_data
```

```python
# Plot average fuel consumption by fuel type and manufacturer
plt.figure(figsize=(12, 8))
sns.barplot(data=average_data, x='Fuel Type', y='OBFCM Fuel consumption (l/100 km)', hue='Manufacturer', palette='viridis')
plt.title('Average Fuel Consumption by Fuel Type and Manufacturer')
plt.xlabel('Fuel Type')
plt.ylabel('Average Fuel Consumption (l/100 km)')
plt.xticks(rotation=45)
plt.legend(title='Manufacturer', loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()

# Plot average CO2 emissions by fuel type and manufacturer
plt.figure(figsize=(12, 8))
sns.barplot(data=average_data, x='Fuel Type', y='OBFCM CO2 emissions (g/km)', hue='Manufacturer', palette='viridis')
plt.title('Average CO2 Emissions by Fuel Type and Manufacturer')
plt.xlabel('Fuel Type')
plt.ylabel('Average CO2 Emissions (g/km)')
plt.xticks(rotation=45)
plt.legend(title='Manufacturer', loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

```python
# Top ten

# Sort the average_data DataFrame by the desired column (e.g., 'OBFCM Fuel Consumption (l/100 km)')
top_ten_fuel_consumption = average_data.sort_values(by='OBFCM Fuel consumption (l/100 km)', ascending=False).head(10)
top_ten_co2_emissions = average_data.sort_values(by='OBFCM CO2 emissions (g/km)', ascending=False).head(10)

# Define a custom palette
custom_palette = sns.color_palette("husl", len(top_ten_fuel_consumption['Manufacturer'].unique()))

# Plot average fuel consumption by fuel type and manufacturer for the top ten values
plt.figure(figsize=(12, 8))
ax = sns.barplot(data=top_ten_fuel_consumption, x='Fuel Type', y='OBFCM Fuel consumption (l/100 km)', hue='Manufacturer', palette=custom_palette)
plt.title('Top Ten Average Fuel Consumption by Fuel Type and Manufacturer')
plt.xlabel('Fuel Type')
plt.ylabel('Average Fuel Consumption (l/100 km)')
plt.xticks(rotation=45)

# Move the legend outside of the plot
plt.legend(title='Manufacturer', loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()

# Plot average CO2 emissions by fuel type and manufacturer for the top ten values
plt.figure(figsize=(12, 8))
ax = sns.barplot(data=top_ten_co2_emissions, x='Fuel Type', y='OBFCM CO2 emissions (g/km)', hue='Manufacturer', palette=custom_palette)
plt.title('Top Ten Average CO2 Emissions by Fuel Type and Manufacturer')
plt.xlabel('Fuel Type')
plt.ylabel('Average CO2 Emissions (g/km)')
plt.xticks(rotation=45)

# Move the legend outside of the plot
plt.legend(title='Manufacturer', loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```

```python
# Least ten

# Sort the average_data DataFrame by the desired column in ascending order
bottom_ten_fuel_consumption = average_data.sort_values(by='OBFCM Fuel consumption (l/100 km)', ascending=True).head(10)
bottom_ten_co2_emissions = average_data.sort_values(by='OBFCM CO2 emissions (g/km)', ascending=True).head(10)

# Define a custom palette
custom_palette = sns.color_palette("husl", len(bottom_ten_fuel_consumption['Manufacturer'].unique()))

# Plot average fuel consumption by fuel type and manufacturer for the least ten values
plt.figure(figsize=(12, 8))
ax = sns.barplot(data=bottom_ten_fuel_consumption, x='Fuel Type', y='OBFCM Fuel consumption (l/100 km)', hue='Manufacturer', palette=custom_palette)
plt.title('Least Ten Average Fuel Consumption by Fuel Type and Manufacturer')
plt.xlabel('Fuel Type')
plt.ylabel('Average Fuel Consumption (l/100 km)')
plt.xticks(rotation=45)

# Move the legend outside of the plot
plt.legend(title='Manufacturer', loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()

# Plot average CO2 emissions by fuel type and manufacturer for the least ten values
plt.figure(figsize=(12, 8))
ax = sns.barplot(data=bottom_ten_co2_emissions, x='Fuel Type', y='OBFCM CO2 emissions (g/km)', hue='Manufacturer', palette=custom_palette)
plt.title('Least Ten Average CO2 Emissions by Fuel Type and Manufacturer')
plt.xlabel('Fuel Type')
plt.ylabel('Average CO2 Emissions (g/km)')
plt.xticks(rotation=45)

# Move the legend outside of the plot
plt.legend(title='Manufacturer', loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```

Measurement Method(OBFCM vs WLTP)

```python

# Select the relevant columns for comparison
comparison_data = car_agg[['Fuel Type', 'OBFCM Fuel consumption (l/100 km)', 'WLTP Fuel consumption (l/100 km)',
                                'OBFCM CO2 emissions (g/km)', 'WLTP CO2 emissions (g/km)']]

# Melt the DataFrame to reshape it for visualization
melted_data = comparison_data.melt(id_vars='Fuel Type', var_name='Measurement Method', value_name='Value')

# Plot fuel consumption comparison
plt.figure(figsize=(12, 8))
sns.barplot(data=melted_data[melted_data['Measurement Method'].str.contains('Fuel consumption')],
            x='Fuel Type', y='Value', hue='Measurement Method')
plt.title('Comparison of Fuel Consumption (OBFCM vs. WLTP)')
plt.xlabel('Fuel Type')
plt.ylabel('Fuel Consumption (l/100 km)')
plt.xticks(rotation=45)
plt.legend(title='Measurement Method')
plt.tight_layout()
plt.show()

# Plot CO2 emissions comparison
plt.figure(figsize=(12, 8))
sns.barplot(data=melted_data[melted_data['Measurement Method'].str.contains('CO2 emissions')],
            x='Fuel Type', y='Value', hue='Measurement Method')
plt.title('Comparison of CO2 Emissions (OBFCM vs. WLTP)')
plt.xlabel('Fuel Type')
plt.ylabel('CO2 Emissions (g/km)')
plt.xticks(rotation=45)
plt.legend(title='Measurement Method')
plt.tight_layout()
plt.show()
```

```python
## Absolute gap

# Group the data by Manufacturer
manufacturer_groups = car_agg.groupby('Manufacturer')

# Calculate mean absolute gap for Fuel Consumption and CO2 Emissions for each Manufacturer
mean_abs_gap_fuel_manufacturer = manufacturer_groups['absolute gap Fuel consumption (l/100 km)'].mean()
mean_abs_gap_co2_manufacturer = manufacturer_groups['absolute gap CO2 emissions (g/km)'].mean()

# Calculate mean absolute gap for Fuel Consumption and CO2 Emissions for each Fuel Type
fuel_type_groups = car_agg.groupby('Fuel Type')
mean_abs_gap_fuel_fuel_type = fuel_type_groups['absolute gap Fuel consumption (l/100 km)'].mean()
mean_abs_gap_co2_fuel_type = fuel_type_groups['absolute gap CO2 emissions (g/km)'].mean()

# Visualize the data
plt.figure(figsize=(14, 6))

# Plot mean absolute gap for Fuel Consumption by Manufacturer
plt.subplot(1, 2, 1)
mean_abs_gap_fuel_manufacturer.plot(kind='bar', color='skyblue')
plt.title('Mean Absolute Gap for Fuel Consumption by Manufacturer')
plt.ylabel('Mean Absolute Gap (l/100 km)')
plt.xlabel('Manufacturer')

# Plot mean absolute gap for CO2 Emissions by Manufacturer
plt.subplot(1, 2, 2)
mean_abs_gap_co2_manufacturer.plot(kind='bar', color='lightgreen')
plt.title('Mean Absolute Gap for CO2 Emissions by Manufacturer')
plt.ylabel('Mean Absolute Gap (g/km)')
plt.xlabel('Manufacturer')

plt.tight_layout()
plt.show()

# Visualize the data for Fuel Type
plt.figure(figsize=(14, 6))

# Plot mean absolute gap for Fuel Consumption by Fuel Type
plt.subplot(1, 2, 1)
mean_abs_gap_fuel_fuel_type.plot(kind='bar', color='skyblue')
plt.title('Mean Absolute Gap for Fuel Consumption by Fuel Type')
plt.ylabel('Mean Absolute Gap (l/100 km)')
plt.xlabel('Fuel Type')

# Plot mean absolute gap for CO2 Emissions by Fuel Type
plt.subplot(1, 2, 2)
mean_abs_gap_co2_fuel_type.plot(kind='bar', color='lightgreen')
plt.title('Mean Absolute Gap for CO2 Emissions by Fuel Type')
plt.ylabel('Mean Absolute Gap (g/km)')
plt.xlabel('Fuel Type')

plt.tight_layout()
plt.show()
```

```python

##Percentage gap

# Group the data by Manufacturer
manufacturer_groups = car_agg.groupby('Manufacturer')

# Calculate mean Percentage gap for Fuel Consumption and CO2 Emissions for each Manufacturer
mean_per_gap_fuel_manufacturer = manufacturer_groups['percentage gap Fuel consumption (%)'].mean()
mean_per_gap_co2_manufacturer = manufacturer_groups['percentage gap CO2 emissions (%)'].mean()

# Calculate mean Percentage gap for Fuel Consumption and CO2 Emissions for each Fuel Type
fuel_type_groups = car_agg.groupby('Fuel Type')
mean_per_gap_fuel_fuel_type = fuel_type_groups['percentage gap Fuel consumption (%)'].mean()
mean_per_gap_co2_fuel_type = fuel_type_groups['percentage gap CO2 emissions (%)'].mean()

# Visualize the data
plt.figure(figsize=(14, 6))

# Plot mean Peercentage gap for Fuel Consumption by Manufacturer
plt.subplot(1, 2, 1)
mean_per_gap_fuel_manufacturer.plot(kind='bar', color='skyblue')
plt.title('Mean Percentage Gap for Fuel Consumption by Manufacturer')
plt.ylabel('Mean Percentage Gap (l/100 km)')
plt.xlabel('Manufacturer')

# Plot mean Percentage gap for CO2 Emissions by Manufacturer
plt.subplot(1, 2, 2)
mean_per_gap_co2_manufacturer.plot(kind='bar', color='lightgreen')
plt.title('Mean Percentage Gap for CO2 Emissions by Manufacturer')
plt.ylabel('Mean Percentage Gap (g/km)')
plt.xlabel('Manufacturer')

plt.tight_layout()
plt.show()

# Visualize the data for Fuel Type
plt.figure(figsize=(14, 6))

# Plot mean Percentage gap for Fuel Consumption by Fuel Type
plt.subplot(1, 2, 1)
mean_per_gap_fuel_fuel_type.plot(kind='bar', color='skyblue')
plt.title('Mean Percentage Gap for Fuel Consumption by Fuel Type')
plt.ylabel('Mean Percentage Gap (l/100 km)')
plt.xlabel('Fuel Type')

# Plot mean Percentage gap for CO2 Emissions by Fuel Type
plt.subplot(1, 2, 2)
mean_per_gap_co2_fuel_type.plot(kind='bar', color='lightgreen')
plt.title('Mean Percentage Gap for CO2 Emissions by Fuel Type')
plt.ylabel('Mean Percentage Gap (g/km)')
plt.xlabel('Fuel Type')

plt.tight_layout()
plt.show()
```

```python
# Correlation between Fuel consumption,CO2 emissions, fuel type


selected_data=['Fuel Type','OBFCM Fuel consumption (l/100 km)','WLTP Fuel consumption (l/100 km)','OBFCM CO2 emissions (g/km)','WLTP CO2 emissions (g/km)']


#selected_data=['Manufacturer','OBFCM Fuel consumption (l/100 km)','WLTP Fuel consumption (l/100 km)','OBFCM CO2 emissions (g/km)','WLTP CO2 emissions (g/km)']



df=car_agg[selected_data]


# Get dummies for Manufacturer and Fuel Type columns
df1 = pd.get_dummies(df, columns=['Fuel Type'])

# Calculate correlation matrix
corr_matrix = df1.corr()



# Mask upper triangular part of the correlation matrix
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Plot triangular heatmap with a larger size
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=1, mask=mask)
plt.title("Correlation Heatmap")
plt.show()
```
